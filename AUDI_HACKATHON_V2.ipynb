```python
def custom_response(user_input):
    for pattern, response in pairs:
        if pattern.lower() == user_input.lower():
            return response
    return "I'm not sure how to answer that. Please try a different question."

def send_message(event=None):
    user_input = user_entry.get()
    user_entry.delete(0, tk.END)
    
    response = custom_response(user_input)

    conversation.insert(tk.END, f"You: {user_input}\n\n")
    conversation.insert(tk.END, f"AudiKi: {response}\n\n")

# Create the main window
root = tk.Tk()
root.title("Chatbot")

# Create a scrollbar
scrollbar = tk.Scrollbar(root)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

# Create a text widget to display the conversation
conversation = tk.Text(root, wrap=tk.WORD, yscrollcommand=scrollbar.set)
conversation.pack()

scrollbar.config(command=conversation.yview)

# Create an entry widget for user input
user_entry = tk.Entry(root)
user_entry.bind("<Return>", send_message)
user_entry.pack()

# Create a send button
send_button = tk.Button(root, text="Send", command=send_message)
send_button.pack()

root.mainloop()
```
